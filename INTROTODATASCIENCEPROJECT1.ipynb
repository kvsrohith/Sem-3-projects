```python
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns',100)
```

```python
#load the data...

df = pd.read_csv('telco-customer-churn.csv')
```

```python
#preview the data

df.head()
```

```python
df.info()
```

```python
df['TotalCharges'] = df['TotalCharges'].astype('float')
```

```python
df['TotalCharges'].value_counts()
```

```python
#alternate solution

df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors='coerce')
```

```python
df.info()
```

# Data Analysis
            -Uniariate
                  -Categorical
                  -Continous

            -BiVariate
                   -Continous-Continous
                   -Continous-Categorical
                   -Categorical-Categorical

```python
df.head()
```

```python
df_1 = df[df['Churn']=='Yes']
df_0 = df[df['Churn']=='No']
```

```python
df_1.head()
```

```python
df_0.head()
```

```python
sns.distplot(df_1['MonthlyCharges'],hist=False,label='Churn')
sns.distplot(df_0['MonthlyCharges'],hist=False,label='Non Churn')
```

```python
df.head()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Churn')
sns.countplot('Contract',data=df_1)
plt.figure(figsize=(10,5))
plt.subplot(1,2,2)
plt.title('Non Churn')
sns.countplot('Contract',data=df_0)
           
```

```python
#BiVariate Analysis
```

```python
# Contract/MonthlyCharges
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Churn')
sns.boxplot(x='Contract',y='MonthlyCharges',data=df_1)
plt.subplot(1,2,2)
plt.title('Non Churn')
sns.boxplot(x='Contract',y='MonthlyCharges',data=df_0)
```

```python
#Contract/ Gender
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Churn')
sns.countplot(x='Contract', hue='gender',data=df_1,palette='Set1')
plt.subplot(1,2,2)
plt.title('Non Churn')
sns.countplot(x='Contract',hue='gender',data=df_0)
```
